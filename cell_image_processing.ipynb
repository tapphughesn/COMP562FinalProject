```python
#you'll need to pip install Pillow and Glob3
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import glob

train_imgs = []
train_labels = []

#keep extracted zip file in same directory as your notebook
for f in glob.glob('blood-cells/dataset2-master/dataset2-master/images/TRAIN/EOSINOPHIL/*'):
    im = np.asarray(Image.open(f))
    train_imgs.append(im)
    train_labels.append(0)

for f in glob.glob('blood-cells/dataset2-master/dataset2-master/images/TRAIN/LYMPHOCYTE/*'):
    im = np.asarray(Image.open(f))
    train_imgs.append(im)
    train_labels.append(1)

for f in glob.glob('blood-cells/dataset2-master/dataset2-master/images/TRAIN/MONOCYTE/*'):
    im = np.asarray(Image.open(f))
    train_imgs.append(im)
    train_labels.append(2)
    
for f in glob.glob('blood-cells/dataset2-master/dataset2-master/images/TRAIN/NEUTROPHIL/*'):
    im = np.asarray(Image.open(f))
    train_imgs.append(im)
    train_labels.append(3)
```
